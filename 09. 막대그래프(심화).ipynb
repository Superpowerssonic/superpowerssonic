## 9. 막대그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
plt.bar(label,values)
plt.show()
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
plt.barh(label,values)
plt.show()
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
plt.barh(label,values)
plt.xlim(180, 200)
plt.show()
```

```python
bar = plt.bar(label, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
plt.plot()
```

```python
bar=plt.bar(label, values)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha="center", color='b')
plt.ylim(180,200)
```
