## 6. 텍스트

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, marker='o')

for idx, txt in enumerate(y): # y데이터로 쓰고 있는 값을 그래프에 표시해준다. enumearte을 통해서 값을 하나씩 가져온다. idx는 0부터 시작하니까 0,1,2 순으로 들어간다
    plt.text(x[idx],y[idx],txt)
```

```python
plt.plot(x,y, marker='o')

for idx, txt in enumerate(x): # y데이터로 쓰고 있는 값을 그래프에 표시해준다. enumearte을 통해서 값을 하나씩 가져온다. idx는 0부터 시작하니까 0,1,2 순으로 들어간다
    plt.text(x[idx],y[idx],txt)
```

```python
plt.plot(x,y, marker='o')

for idx, txt in enumerate(y): # y데이터로 쓰고 있는 값을 그래프에 표시해준다. enumearte을 통해서 값을 하나씩 가져온다. idx는 0부터 시작하니까 0,1,2 순으로 들어간다
    plt.text(x[idx],y[idx] + 0.1,txt, ha='center', color='blue')
```
