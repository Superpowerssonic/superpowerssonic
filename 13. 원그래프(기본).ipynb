## 13. 원그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
values =[30,25,13,20,30]
labels = ['python','java','javascript','c','etc']
plt.pie(values)
plt.show()
```

```python
values =[30,25,13,20,30]
labels = ['python','java','javascript','c','etc']
# plt.pie(values, labels=labels, autopct='%.2f')  # 1f - 소숫점 1자리까지 
plt.pie(values, labels=labels, autopct='%.2f%%', startangle=90, counterclock=False)  # 1f - 소숫점 1자리까지 
plt.show()
```

```python
values =[30,25,13,20,30]
labels = ['python','java','javascript','c','etc']
explode = [0.05] *5
plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2,0.3), title='언어별 선호도')
plt.show()
```
