##  Covid-19 백신 종류별 접종 인구

```python
days = [1,2,3]
az = [2,4,8]
fz = [5,1,3]
md = [1,2,5]

plt.plot(days, az)
plt.plot(days, fz)
plt.plot(days, md)
```

```python
plt.plot(days, az, label='az', marker='v', linestyle='--')
plt.plot(days, fz, label='fz', marker='o', linestyle='-.')
plt.plot(days, md, label='md', marker='x')

plt.legend()
```
