## 14. 원그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
values =[30,25,13,20,30]
labels = ['python','java','javascript','c','etc']
# plt.pie(values, labels=labels, autopct='%.2f')  # 1f - 소숫점 1자리까지 
colors = ['blue','red','yellow','green','pink']
explode = [0.05] * 5
plt.pie(values, labels=labels, autopct='%.2f%%', counterclock=False, colors=colors, explode=explode)  # 1f - 소숫점 1자리까지 
plt.show()
```

```python
wedgeprops={'width': 0.7}  # wedgeprops는 딕셔너리 형태로 {} 사용
plt.pie(values, labels=labels, autopct='%.2f%%', counterclock=False, colors=colors, explode=explode,wedgeprops=wedgeprops)  # 1f - 소숫점 1자리까지 
plt.show()
```

```python
wedgeprops={'width': 0.7, 'edgecolor':'w', 'linewidth':3}  # wedgeprops는 딕셔너리 형태로 {'':''} 사용
plt.pie(values, labels=labels, autopct='%.2f%%', counterclock=False, colors=colors,wedgeprops=wedgeprops)  # 1f - 소숫점 1자리까지 
plt.show()
```

```python
def custom_autopct(pct):
    return('%.2f%%' % pct) if pct >= 10 else''

plt.pie(values, labels=labels, autopct=custom_autopct, counterclock=False, colors=colors,wedgeprops=wedgeprops)  # 1f - 소숫점 1자리까지 
plt.show()
```

```python
def custom_autopct(pct):
#     return '{:.1f}%'.format(pct) if pct >=10 else'' # .1f 는 소숫점 1번째까지
    return '{:.0f}%'.format(pct) if pct >=10 else'' #.f 는 정수부분만 보여준다


plt.pie(values, labels=labels, autopct=custom_autopct, counterclock=False, colors=colors,wedgeprops=wedgeprops, pctdistance=0.8)  # 1f - 소숫점 1자리까지 
plt.show()
```
