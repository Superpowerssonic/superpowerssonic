## 8. 막대그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
plt.bar(label,values)
plt.show()
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
plt.bar(label,values, color='r')
plt.show()
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
colors =['r','g','b']  # 각각의 색깔 정의
plt.bar(label,values, color=colors, alpha=0.5) # alpha =int 투명도
plt.show()
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
plt.bar(label,values)
plt.ylim(180,200)
plt.show()
```

```python
plt.bar(label,values, width=0.3) # width=int 는 막대기 두꼐
plt.ylim(180,200)
plt.show()
```

```python
plt.bar(label, values, width=0.3)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
label =['강백호','서태웅','정대만']
values = [190, 187, 195]
ticks = ['1번','2번','3번']

plt.bar(label, values)
plt.xticks(label, ticks, rotation=90)  # label 값 대신 ticks로 정의한 값을 사용하고 싶을 경우
```
