## 15. 산점도 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
import pandas as pd
df = pd.read_excel("../pandas/score.xlsx")
df
```

```python
df['학년']=[3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어 점수')
plt.ylabel(';수학 점수')
        
```

```python
import numpy as np
sizes = np.random.rand(8) * 100
sizes
```

```python
plt.scatter(df['영어'],df['수학'], s = sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
sizes =df['학년'] * 500 
sizes
```

```python
plt.scatter(df['영어'],df['수학'], s = sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.scatter(df['영어'],df['수학'], s = sizes, c=df['학년'],cmap='viridis', alpha=0.3)  # ggl matplotlib cmap 검색
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df['영어'],df['수학'], s = sizes, c=df['학년'],cmap='viridis', alpha=0.3)  # ggl matplotlib cmap 검색
plt.xlabel('영어점수')
plt.ylabel('수학점수')
plt.colorbar(ticks=[1,2,3], label='학년', shrink=0.5, orientation='horizontal')
```
