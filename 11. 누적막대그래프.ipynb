```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
df = pd.read_excel('../pandas/score.xlsx')
df
```

```python
plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'])
```

```python
plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'], bottom=df['국어'])
```

```python
# plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'], bottom=df['국어'])
```

```python
plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'], bottom=df['국어'])
plt.bar(df['이름'],df['수학'], bottom=df['국어']+df['영어'])
```

```python
plt.bar(df['이름'],df['국어'],label='국어')
plt.bar(df['이름'],df['영어'], bottom=df['국어'],label='영어')
plt.bar(df['이름'],df['수학'], bottom=df['국어']+df['영어'], label='수학')
plt.xticks(rotation=60)

plt.legend(loc='upper right')
```
