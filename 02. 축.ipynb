# 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title("꺽은선 그래프", fontdict={'family':'HYGungSo-Bold', 'size':20})
```

```python
plt.plot(x,y)
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red')
plt.ylabel('y축', color='#00aa00') # rgb로 green
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red', loc='right')
plt.ylabel('y축', color='#00aa00', loc='top') # rgb로 green
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
