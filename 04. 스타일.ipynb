## 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
# matplotlib.rcParams['font.family'] = 'Apple Gothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 음수가 깨지는 현상 방지
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5) # linewidth=int 선 두께
```

## marker

```python
plt.plot(x,y, marker='x')
```

```python
plt.plot(x,y, marker='x', linestyle='None')
```

```python
plt.plot(x,y, marker='x', markersize=10) # ggl matplotlibmarker 검색 하면 마커 형태 확인 가능
```

```python
plt.plot(x,y, marker='o',markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--', marker='o', markerfacecolor='yellow', markeredgecolore='red',)
```

```python
plt.plot(x,y, linestyle='-.', marker='o', markerfacecolor='yellow', markeredgecolor='red') #ggl matplotlib linestyles 검색
```

```python
plt.plot(x,y, linestyle=':', marker='o', markerfacecolor='yellow', markeredgecolor='red') #ggl matplotlib linestyles 검색
```

```python
plt.plot(x,y, linestyle='-', marker='o', markerfacecolor='yellow', markeredgecolor='red') #ggl matplotlib linestyles 검색
```

## 포멧

```python
plt.plot(x,y, 'ro--') # 컬러, 마커형태, 라인스타일 ro--
```

```python
plt.plot(x,y, 'bo:') # 컬러, 마커형태, 라인스타일 ro--
```

```python
plt.plot(x,y, 'go') # 컬러, 마커형태, 라인스타일 ro--
```

```python
plt.plot(x,y, color='green', linestyle='None', marker='o')
```

## 축약어

plt.plot(x,y, marker='o',mfc='green', ms=10, mec='red', ls='--') #matplotlib markerfacecolore mfc 로 검색

## 투명도

```python
plt.plot(x,y, marker='o',mfc='green', ms=10, mec='red', ls='--', alpha=0.3) #alpha는 투명도
```

## 그래프 크기

```python
plt.figure(figsize=(10,5)) # ((가로,세로))
plt.plot()
```

```python
plt.figure(figsize=(5,10)) # ((가로,세로))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=200) # dpi=해상도 dpi = dot per inch
plt.plot(x,y)
```

## 그래프 자체 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor='a1c3ff')
plt.plot(x,y)
```
